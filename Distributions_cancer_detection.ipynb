```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plot
%matplotlib inline
import seaborn as sns
print("Loading done")
```

```python
cancer_filepath = "./cancer.csv"

cancer_data = pd.read_csv(cancer_filepath, index_col='Id')

cancer_data.head()
```

```python
# Get the 'Area (mean)' comparison for the different diagnosis

plot.figure(figsize=(10,10))
sns.histplot(data=cancer_data, x=cancer_data['Area (mean)'], hue=cancer_data['Diagnosis'])
```

```python
# KDE plots for benign and malignant tumors
plot.figure(figsize=(10,8))
sns.kdeplot(data=cancer_data, x='Radius (worst)', fill=True, hue='Diagnosis')
```
