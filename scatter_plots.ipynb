```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup done")
```

```python
candy_filepath = "./candy.csv"

candy_data = pd.read_csv(candy_filepath, index_col="id")
```

```python
candy_data.head()
```

```python
# Draw a scatter plt showing the relationship between 'sugarpercent' and 'winpercent'
plt.figure(figsize=(12,8))

sns.scatterplot(x='sugarpercent', y='winpercent', data=candy_data)
```

```python
# Add a regression line in the scatter pot
plt.figure(figsize=(12,10))
sns.regplot(x='sugarpercent', y='winpercent', data=candy_data)
```

```python
# Scatter plot showing the relationship between 'pricepercent', 'winpercent', and 'chocolate'
plt.figure(figsize=(12,10))
sns.scatterplot(data=candy_data, x='pricepercent', y='winpercent', hue='chocolate')
```

```python
# Use regression lines to deducce the relationship between chocolate candies and those without chocolate
plt.figure(figsize=(12,8))
sns.lmplot(data=candy_data, x='pricepercent', y='winpercent', hue='chocolate')
```

```python
# Create a categorical scatter plot to highlight the relationship between 'chocolate' and 'winpercent'
plt.figure(figsize=(12,10))
sns.swarmplot(x=candy_data['chocolate'], y=candy_data['winpercent'])

```
