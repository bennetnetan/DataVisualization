```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup done")
```

```python
ign_filepath = "./ign_scores.csv"

ign_data = pd.read_csv(ign_filepath, index_col="Platform")
```

```python
# preview the dataset
ign_data
```

```python
# Draw a bargraph giving the average of the racing games
plt.figure(figsize=(12,7))

plt.title("Average Of Racing Games Across Multiple Platforms")

sns.barplot(x=ign_data.index, y=ign_data['Racing'])
# Adding labels and title to the x-axis and y-axis respectively
plt.xticks(rotation=45)
plt.ylabel('Avg Rating')
```

```python
# Use a heatmap to get the average game score by platform and genre
plt.figure(figsize=(12,7))

plt.title("Average Game Score By Platform And Genre")

sns.heatmap(ign_data, annot=True)

plt.xlabel("Genre")

plt.xticks(rotation=45)
```
